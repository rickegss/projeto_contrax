```python
import pandas as pd
import numpy as np

df = pd.read_csv("1. CONTRATOS.csv", sep=";", encoding="utf-8")

df = df.drop(df.columns[0], axis=1)
df = df.drop(columns=["Parcela", "Nr Contrato", "Anexos", "Prestador", "CNPJ", "Cod Datasul", "Nr Pedido"], axis=1)
columns = ["ano", "mes", "data_lancamento", "data_emissao", "data_vencimento", "tipo", "contrato", "referente", "documento", "classificacao", "estabelecimento", "situacao", "status", "valor"]
df.columns = columns
df.head()
```

```python
df["ano"] = df["ano"].astype(int)
```

```python
mes_dict = {
    "jan": 1, "fev": 2, "mar": 3, "abr": 4,
    "mai": 5, "jun": 6, "jul": 7, "ago": 8,
    "set": 9, "out": 10, "nov": 11, "dez": 12
}

df["mes"] = df["mes"].map(mes_dict)
df.head()
```

```python
date_columns = ["data_lancamento", "data_emissao", "data_vencimento"]

for date_column in date_columns:
    df[date_column] = pd.to_datetime(df[date_column], format="%d/%m/%Y", errors="coerce")
    df[date_column] = df[date_column].where(df[date_column].notna(), None)
```

```python
df["valor"] = pd.to_numeric(df["valor"].str.replace(".", "").str.replace(",", ".").replace(" - ", np.nan), errors='coerce')
```

```python
df.to_csv("parcelas.csv", index=False, encoding="utf-8")
```
