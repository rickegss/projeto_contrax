```python
import pandas as pd 

df = pd.read_csv("contratos.csv", encoding="utf-8", sep=";")
df = df.drop(['Cod Prest', 'Op/Proj', 'Endereço', 'Bairro', 'Cidade/UF', 'CEP', 'Contato', 'Telefones', 'Email', 'I.Municipal', 'I.Estadual', 'Obs', 'Renovação', 'URL', 'Usuário', 'Senha', 'Prestador'], axis=1)

columns = ["situacao",  "numero", "contrato", "conta", "centro_custo", "estabelecimento", "classificacao", "descricao", "cnpj", "anexos", "valor_contrato", "inicio", "termino"]
df.columns = columns
df.head()
```

```python
df["valor_contrato"] = df["valor_contrato"].str.replace(".", "").str.replace(",", ".").astype(float, errors="ignore") 
```

```python
date_columns = ["inicio", "termino"]

for column in date_columns:
    df[column] = pd.to_datetime(df[column], format="%d/%m/%Y", errors="coerce")
```

```python
df.to_csv("contratos1.csv", index=False, encoding="utf-8")
```
